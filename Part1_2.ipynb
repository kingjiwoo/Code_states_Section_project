## 1.1 분류를 위한 베이스라인 정확도를 계산하는 `calculate_baseline_score` 함수를 작성하세요.
- 예측해야 하는 target은 `shot_made_flag` 입니다. 
- 전체 데이터를 써서 baseline **accuracy score**를 계산하세요.
- 최빈 클래스를 사용하는 baseline을 가정하세요.

```python

def calculate_baseline_score(df):
    accuracy_score = 0
    
    # YOUR CODE HERE
    from sklearn.metrics import accuracy_score
    baseline = df['shot_made_flag'].mode()
    pred = [baseline] * len(df)
    accuracy_score = accuracy_score(pred, df['shot_made_flag'])

    
    # YOUR CODE ENDS HERE
    
    return accuracy_score
```

```python
baseline_accuracy_score = calculate_baseline_score(df)
print('baseline accuracy score:', baseline_accuracy_score)
```

## 1.2 데이터셋을 아래의 기준을 사용해서 train/test 셋으로 나누는 `split_data` 함수를 작성하세요.

- `test set`: **2014-15** ~ **2015-16** 시즌
- NBA 시즌은 **10월**에 시작해서 **6월**에 종료합니다. 

```python
from sklearn.model_selection import train_test_split

def split_data(df):
    train = None
    test = None

    # YOUR CODE HERE
    con1 = df['game_date'].str.startswith('2015')
    con2 = df['game_date'].str.startswith('2014-1')
    con3 = df['game_date'].str.startswith('2016-0')
    test = df.loc[(con1|con2|con3),:]
    train =  df.drop(test.index)


    
    # YOUR CODE ENDS HERE
    
    return train, test
```

```python
train, test = split_data(df.copy())
```

## 1.3 새로운 feature를 만드는 `feature_engineering` 함수를 작성하세요.

- 최소 4개 이상의 새로운 feature를 만드세요. 아래의 리스트를 사용하셔도 되고 스스로의 아이디어로 만들어 주셔도 괜찮습니다.
    - **Homecourt Advantage**: 홈 팀(`htm`)이 Los Angeles Lakers (`LAL`)인지 여부
    - **Opponent**: LA Lakers를 상대하는 팀을 나타내는 feature
    - **Seconds remaining in the period**: 피리어드에 남은 총 시간 (분과 초를 합칩니다)
    - **Seconds remaining in the game**: 게임에 남은 총 시간 (피리어드와 초를 합칩니다). 농구에서 게임은 총 4개의 피리어드로 이루어져 있습니다 (각 12분)
    - **Made previous shot**: 코비 브라이언트의 예전 슛이 성공했는지 여부
    

```python
def feature_engineering(df):
    df_tmp = df.copy()

    # YOUR CODE HERE
    df_tmp['Seconds remaining in the period'] = (df_tmp['minutes_remaining'] * 60) + (df_tmp['seconds_remaining']) #1
    df_tmp['Seconds remaining in the game'] = (4 - df_tmp['period'])*12 + df_tmp['Seconds remaining in the period'] #2
    #3
    def f(df_tmp):
        if df_tmp['htm'] == 'LAL':
            return 1
        else:
            return 0
    df_tmp['HomecourtAdvantage'] = df_tmp.apply(f, axis = 1)
    #4
    def f1(df_tmp):
        if df_tmp['htm'] == 'LAL':
            return df_tmp['vtm']
        else:
            return df_tmp['htm']
    df_tmp['Opponent'] = df_tmp.apply(f1, axis = 1)
    # YOUR CODE ENDS HERE
    
    return df_tmp
```

```python
train_processed = feature_engineering(train)
test_processed = feature_engineering(test)
```

## 1.4 train / test 셋을 features(X)와 target(y)으로 각각 나누는 `divide_data` 함수를 작성하세요.

```python
def divide_data(df):
    X = None
    y = None

    # YOUR CODE HERE

    y = df['shot_made_flag']
    X = df.loc[:,['action_type', 'shot_type',
       'shot_zone_range', 'shot_distance',
       'loc_x', 'loc_y','Seconds remaining in the period',
       'Seconds remaining in the game', 'HomecourtAdvantage']]
    
    # YOUR CODE ENDS HERE

    return X, y
```

## 2.1 Decision Tree 또는 RandomForest 모델 학습 파이프라인을 생성하고 학습하는 `fit_trees` 함수를 작성하세요. 
- Scikit-learn의 `pipeline`을 활용하여 범주형 feature의 encoder, imputer, 모델을 포함하는 하나의 학습 파이프라인을 구축하세요.
- 구축한 파이프라인에 대해 `RandomizedSearchCV`를 사용해 CV를 진행하세요.
    - 적어도 두 개 이상의 하이퍼파라미터에 대해 탐색을 진행하세요.
    - F1 score을 최대화하는 하이퍼파라미터 조합을 선택하세요.
- RandomForestClassifier, RandomizedSearchCV 등 랜덤성이 포함되는 방법론에 대해 모두 `random_state=42`로 하세요.
- cv는 5 이하, n_iter은 10회 이하로 하세요.

```python
import sklearn
from sklearn.pipeline import make_pipeline, Pipeline
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import f1_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.impute import SimpleImputer
from category_encoders import OrdinalEncoder, OneHotEncoder
### IMPORT ANY


def fit_trees(X_train, y_train):
    pipe = None
    clf = None
    
    # YOUR CODE HERE
    pipe = make_pipeline(
    OrdinalEncoder(), 
    SimpleImputer(), 
    RandomForestClassifier(random_state=42)
)
    dists = { 
    'randomforestclassifier__max_depth': range(1, 10), 
    'randomforestclassifier__min_samples_leaf': [1,2],
    'randomforestclassifier__min_samples_split': [1,2],
    'randomforestclassifier__n_estimators': [1000, 1500, 2000]
}


    clf = RandomizedSearchCV(
        pipe, 
        param_distributions=dists, 
        n_iter=20,  
        cv=4, 
        scoring='f1', 
        verbose=0, 
        n_jobs=-1,
        random_state = 42,
        refit = True
    )

    clf.fit(X_train, y_train)
    # YOUR CODE ENDS HERE
    
    return clf
```

```python
clf = fit_trees(X_train, y_train)
print("Best CV F1", clf.best_score_)
print("Best CV params", clf.best_params_)
```

## 2.2 학습된 모델로부터 feature importance를 계산하는 `get_feature_importances` 함수를 작성하세요. 
- feature_importances 변수에 RandomForest / Decision Tree의 feature_importances_ 값을 할당하면 됩니다.

```python
import numpy as np

def get_feature_importances(pipe, feature_names):
    feature_importances = np.zeros(len(feature_names))

    # YOUR CODE HERE
    
    feature_names = pipe.named_steps['ordinalencoder'].get_feature_names()
    feature_importances = pipe.named_steps['randomforestclassifier'].feature_importances_
    
    # YOUR CODE ENDS HERE
    
    return sorted(list(zip(feature_names, feature_importances)), key=lambda x: x[1], reverse=True)
```

```python
feature_importances = get_feature_importances(clf.best_estimator_, list(X_train.columns))
for name, importance in feature_importances:
    print(f"{name}: {importance:.4f}")
```
